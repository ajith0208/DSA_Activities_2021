**1. Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('diabetes_model.csv')
data[['Glucose','Age']][45:61]
```

**2. How many patients in the data are having the blood pressure level above 120?**

```python
bp = (data['BloodPressure']>120)
print('No. of patients having the Blood Pressure level above 120 : ',(bp==True).sum())
```

**3. Select the rows the age is between 24 and 29 (inclusive)**

```python
data[data['Age'].between(24,29)]
```

**4. Find the minimum and maximum value of insulin level in the data.**

```python
insulin = data['Insulin']
print('Minimum value of insulin level in the data : ',min(insulin))
print('Maximum value of insulin level in the data : ',max(insulin))
```

**5. Create a histogram of "Age".**

```python
plt.figure(figsize=(8,8))
plt.hist(data['Age'],color='blue',rwidth=.6)
plt.title('Histogram of Age',fontsize=16)
plt.xticks()
plt.yticks()
```

**6. Create a scatter plot of "BMI" vs "Glucose"**

```python
plt.figure(figsize=(8,8))
plt.scatter(data['BMI'],data['Glucose'],s=20,c='red')
plt.title('BMI vs Glucose',fontsize=16)
plt.xlabel('BMI')
plt.ylabel('Glucose')
```

**7. Replace the min value of "BMI" with mean value of BMI**

```python
mean_value = round(data['BMI'].mean(),2)
print('Mean vale of BMI :',mean_value)
print('Minimum value before replace : ',data['BMI'].min())
data[data['BMI']==data['BMI'].min()]=mean_value
print('Minimum value after replace : ',data['BMI'].min())
```

**8. Draw a grouped boxplot with "Outcome" on x-axis and "Glucose" on the y-axis such that we see distributions of Glucose separately for each outcome.**

```python
sns.boxplot(x='Outcome',y='Glucose',data=data)
```

### 5.

**-** Create a numpy array (first) with 5 x 5 size and all elements are 0 in that.

**-** Create a numpy array (second) with 5 x 5 and all elements are 1 in that.

**-** Combine the arrays first and second vertically, the final array size should be 10 x 5

```python
array_1 = np.zeros((5,5))
array_2 = np.ones((5,5))
array_3 = np.concatenate([array_1,array_2])
print('First Array : ','\n',array_1)
print('\n','Second Array : ','\n',array_2)
print('\n','Combined Array : ','\n',array_3)
```
